```python
import numpy as np
import pandas as pd
import torch
from torch import cuda
from transformers import AutoTokenizer
from datasets import load_dataset, load_metric
from transformers import AutoModelForSeq2SeqLM
```

```python
#Load sample data collected from test data set that will can be used in demo
sampledata = pd.read_csv('DEMO_Files/sampletestdata.csv')
sampledata.head()
```

```python
#As our model was finetuned on t5-small pretrained model
model_checkpoint = "t5-small"

#from hugging face library import the corresponding tockenizer for this model
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint)

#as task is to summarize T5 model expects this prefix
prefix = "summarize: "

#if needed load rough score metric library
metric = load_metric("rouge")

#define the limits used while training
max_input_length = 1024
max_target_length = 184

#Load the finetuned model
#load and keep it ready so that it can avoid each time loading
model = AutoModelForSeq2SeqLM.from_pretrained(model_path)
model.eval()
print('model loaded successfully..')
```

```python
#write a function that takes the document text as input
# and returns the summary (Abstract) by the fine tuned model 
def getsummary(document):
    with torch.no_grad():

      text = prefix+str(document)
      text =  ' '.join(text.split())
      
      inputs = tokenizer(text, max_length= max_input_length, pad_to_max_length=True,truncation=True,return_tensors="pt").input_ids
        
      outputs = model.generate(inputs,max_length=max_target_length,do_sample=False)
      
      #Decode the generated token ids back into text:
      preds = tokenizer.decode(outputs[0], skip_special_tokens=True,clean_up_tokenization_spaces=True)
    
    
    return preds
```

```python
import gradio as gr

## Load document, abstract from sample dataset
def load_Document(x):

    #extract row number from drop down list
    id = int(str(x).split('row')[1])
    id = id-1 #index starts from zero in dataframe

    #Get article, abstract from sample dataset
    article = sampledata.loc[id]['article']
    abstract = sampledata.loc[id]['abstract']

    # return document, abstract and empty string to clear the generate summary text field
    return article,abstract,''

#Generate Summary from the first tab
#by calling getsummary method above which generates summary using finuted T5-small model
def Generate_Summary(x):
    return getsummary(x)


#Generate Summary from second tab and update UI elements
#by calling getsummary method above which generates summary using finuted T5-small model
def Generate_Summary_FreeText(x):
    return getsummary(x)

### UI Part of coding
with gr.Blocks() as demo:
    gr.Markdown("### DEMO of Generating Abstract of research papers(arXiv) using pretrained transformer")
    with gr.Tab("Load sample input data From Test data set"):
        dropdown_input = gr.Dropdown(
            ["row1", "row2", "row3","row4","row5"], label="Test Dataset", info="Choose a document from Test dataset")
        text_Document_output = gr.Textbox(max_lines=200,label='Document Text')
        text_GeneratedSummary_output = gr.Textbox(max_lines=100,label = 'Generated Abstract')
        text_button = gr.Button("Generate Abstract/Summary")
        with gr.Accordion("Load Target Abstract",open = False):
          text_Abstract_output = gr.Textbox(max_lines=100, label='')

    with gr.Tab("Free text - summarization "):
        #with gr.Row():
        text_free_input = gr.Textbox(max_lines=200,label='Document')
        text_free_output = gr.Textbox(max_lines=100,label='Summary')
        free_button = gr.Button("Generate Summary")

    
    
    dropdown_input.change(load_Document, inputs =dropdown_input,outputs = [text_Document_output,text_Abstract_output,text_GeneratedSummary_output])
    text_button.click(Generate_Summary, inputs=text_Document_output, outputs=text_GeneratedSummary_output)
    free_button.click(Generate_Summary_FreeText, inputs=text_free_input, outputs=text_free_output)


#Launch the app
demo.launch()
```
